## gestire valori null

il null è un problema per le analisi. indica valori mancanti, o dati incompleti.

la funzione `isnull` identifica se il valore è un null (o NaN)

```python
df_studenti.isnull()
```

```python
df_studenti.isnull().sum()
```

```python
df_studenti[df_studenti.isnull().any(axis=1)]
```

il dataframe recuperato contiene dei NaN (not a number)

quindi adesso come li gestiamo?
- input di nuovi valori?
- eliminiamo le righe?
- altre soluzioni?

```python
#per riempire con nuovi valori possiamo usare la funzione fillna
df_studenti.StudyHours = df_studenti.StudyHours.fillna(df_studenti.StudyHours.mean())
# sostituiamo StudyHours con i cambiamenti di fillna

df_studenti[df_studenti.isnull().any(axis=1)] # controlliamo il risultato
```

```python
# per eliminare delle righe con valori vuoti usiamo dropna
df_studenti = df_studenti.dropna(axis=0, how='any')
df_studenti
```

## Esploriamo i dati
 adesso che abbiamo pulito i dati, possiamo esplorare meglio il DataFrame.

```python
# cominciamo con un confronto tra StudyHours e Grade

#media voti
media_voti = df_studenti.Grade.mean()
#media ore di studio
media_ore = df_studenti['StudyHours'].mean()
#stampa risultati
print('Media ore di studio: {:.2f}\nMedia voti: {:.2f}'.format(media_ore, media_voti))
print(df_studenti[['Grade', 'StudyHours']].mean())
#filtra il dataframe per studenti che hanno studiato più della media
df_studenti_maggioremedia = df_studenti[df_studenti.StudyHours  >= media_ore]
df_studenti_maggioremedia
```

```python
df_studenti_maggioremedia.Grade.mean()
```

Supponiamo che il voto minimo per passare è 60.

Vogliamo aggiungere una colonna per identifare chi ha passato e chi no.

```python
# creiamo una nuova Serie con un indicatore se lo studente ha superato l'esame.
passes = pd.Series(df_studenti.Grade >= 60)
# passes = passes.rename('Pass')
passes.rename('Pass', inplace=True) #inplace fa l'operazione sull'oggetto di partenza, senza bisogno di riassegnare

# aggiungiamo la serie al dataframe come nuova colonna
df_studenti = pd.concat([df_studenti, passes], axis=1)
df_studenti
```

```python
# possiamo fare operazioni analoghe a quelle di un database:
# raggruppa per Pass e conta quanti studenti ci sono per ogni valore
df_studenti.groupby(df_studenti.Pass).Name.count()
```

```python
#possiamo anche aggregare per più colonne [[colonna1, colonna2...]]
df_studenti.groupby(df_studenti.Pass)[['StudyHours', 'Grade']].mean()
```

```python
#possiamo anche ordinare i dati con la funzione sort_values
df_studenti.sort_values('Grade', ascending=False)
# come sempre il risultato è un DataFrame
```

## Introduzione a analisi statistica

```python
df_studenti['Grade'].hist()
```

```python
#ricreo il grafico in matplotlib
var_data = df_studenti['Grade']

fig = plt.figure(figsize=(10,4))
plt.hist(var_data)
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

#aggiungo altre informazioni:
min_val = var_data.min()
max_val = var_data.max()
mean_val = var_data.mean()
med_val = var_data.median()
mode_val = var_data.mode()[0]

plt.axvline(x=min_val, color='gray', linewidth=0.5)
plt.axvline(x=max_val, color='blue', linewidth=0.5)
plt.axvline(x=mean_val, color='yellow', linewidth=0.5)
plt.axvline(x=med_val, color='orange', linewidth=0.5)
plt.axvline(x=mode_val, color='cyan', linewidth=0.5)

fig.show()
```
